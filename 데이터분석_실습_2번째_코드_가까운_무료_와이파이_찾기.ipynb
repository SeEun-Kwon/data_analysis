```python
# 사용할 라이브러리 임포트
import folium
import pandas as pd
import plotly.express as px
```

```python
# 데이터셋 불러오기
data = pd.read_csv('/content/와이파이dataset.csv', encoding='EUC-KR')
data.head()
```

```python
# null 데이터가 얼마나 있는지 확인하기
data.isna().sum() # isna는 비었는지 확인
```

```python
data['설치시설구분'].unique()    # unique 함수: 유일(고유)한 값 추출
```

```python
# 서민.복지시설, 서민/복지시설 -> 서민복지시설 데이터로 전처리
# data.설치시설구분 == data['설치시설구분']

data.loc[data.설치시설구분.str.contains('서민'), '설치시설구분'] = '서민복지시설'  # type object에서 str로 변환
data.설치시설구분.unique()
```

```python
# 설치시설별 무료 wifi 설치 현황수
data['설치시도명'].value_counts()  # unique 값들에 몇개씩 있는지 확인
```

DataFrame.groupby(by=['기준']) <br>
groupby 연산자를 사용해 집단, 그룹별로 데이터를 집계 및 요약 <br>
다양한 변수를 가진 데이터셋을 분석하는데 있어 그룹별로 데이터를 집계하는 분석은 사용빈도가 높아 알아두면 유용!

```python
group_data = data.groupby(by=['설치시도명', '설치시설구분', '서비스제공사명']).count()
group_data
```

```python
# 설치시도별, 설치시설구분, 서비스제공사명 별 무료 wifi 설치 현황
group_data = data.groupby(by=['설치시도명', '설치시설구분', '서비스제공사명', '설치년월']).size().reset_index(name='설치수')
group_data
```

```python
group_data.loc[group_data.서비스제공사명.str.contains('LG') | group_data.서비스제공사명.str.contains('엘지'), '서비스제공사명'] = 'LGU+'
# or 연산
group_data
```

```python
# 내 위치 좌표 찾기
import folium

m = folium.Map(location=[36.3511, 127.3866], zoom_start=13)

# ClickForMarker 함수 적용
m.add_child(folium.ClickForMarker(popup='point'))

# 위도/경도 팝업 활성화되도록 적용 -> 지도를 탐색해서 좌표값을 찾는데 도움
m.add_child(folium.LatLngPopup())
```

```python
# 사용자에게 주소를 받아서 -> 좌표 변환
# 도로명주소를 입력받고 좌표값으로 변환하는 함수를 사용한다.
from geopy.geocoders import Nominatim

def geocoding(address):
  geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
  geo = geolocoder.geocode(address)
  crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}

  return crd

address = input('도로명 주소를 입력하세요 : ')
crd = geocoding(address)
print(crd['lat'])
print(crd['lng'])

myhome = folium.Map(location=[crd['lat'], crd['lng']], zoom_start=14)

folium.Marker([crd['lat'], crd['lng']], icon=folium.Icon(color='red', icon='glyphicon glyphicon-home')).add_to(myhome)

myhome
```

나랑 가장 가까운 무료 와이파이 위치 찾기

```python
from geopy.distance import geodesic
```

```python
my_wifi = pd.DataFrame(columns=['설치시군구명', '설치시설구분', '위도', '경도', '거리'])
myhome = (37.2352614, 127.0754885)

for n in data.index :
  wifi_loc = (data.loc[n, '위도'], data.loc[n, '경도']) # tuple 형태로 구성

  # my_wifi dataframe에 담기
  my_wifi.loc[n] = [data.loc[n, '설치시군구명'], data.loc[n, '설치시설구분'], data.loc[n, '위도'], data.loc[n, '경도'], geodesic(myhome, wifi_loc).kilometers]

my_wifi.head()
```

```python
# 내 위치와 가장 가까운 무료 wifi 10곳 담는다
my_wifi = my_wifi.sort_values(by=['거리']).head(10)   # 오름차순으로 정렬
my_wifi
```

```python
my_map = folium.Map(location=[37.2352614, 127.0754885], zoom_start=14)

for n in my_wifi.index :
  folium.Marker([my_wifi.loc[n, '위도'], my_wifi.loc[n, '경도']],
                popup='<pre>'+my_wifi.loc[n,'설치시설구분']+'</pre>',
                icon=folium.Icon(icon='wifi', prefix='fa')).add_to(my_map)    # marker추가할 때 무조건 add_to 넣어야 my map에 저장됨

folium.Marker([37.2352614, 127.0754885], icon=folium.Icon(color='red', icon='glyphicon glyphicon-home')).add_to(my_map)
my_map
```
