# **13. 함수 적용**

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['학교'] = df['학교'] + '등학교'
# df['키'] = df['키'] + 'cm'
df
```

```python
# 데이터에 함수 적용 (apply)
def add_cm(height):
  return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm)  # 키 데이터에 대해서 add_cm 함수를 호출한 결과 데이터를 반
df
```

```python
def capitalize(lang):
  if pd.notnull(lang):  # NaN이 아닌지
    return lang.capitalize()  # 첫 글자는 대문자로, 나머지는 소문자로

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```
