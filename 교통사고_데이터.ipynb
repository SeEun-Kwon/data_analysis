```python
import pandas as pd
import plotly.express as px
```

데이터 준비하기

```python
df = pd.read_csv('/content/도로교통공단_사망 교통사고 정보_20221231.csv', encoding='euc-kr')
df.head()
```

```python
df.info()   #빈 데이터 없음 확인가능
```

```python
# 1. 발생년월일시 object -> string (문자열 타입)
df = df.astype({'발생년월일시' : 'string'})
df.info()
```

```python
# 2. 발생년월일시 -> 발생시간이라는 컬럼 생성 + 시간을 넣어주기
# 2022-01-01 10:00

df['발생시간'] = df['발생년월일시'].str[11:13]
df = df.astype({'발생시간' : 'int64'})
df.info()
```

```python
# 3. 발생년월일시 -> 날짜타입을 변환

df['발생년월일시'] = pd.to_datetime(df['발생년월일시'].str[:10], format='%Y-%m-%d', errors='raise') # errors: 빈데이터를 그냥 넘어가게해주는 param
df.info()
```

교통사고 현황 지도 시각화

```python
s_df = df[df['발생지시도'] == '서울']
s_df.info()
```

```python
# 시간대별 교통사고 사망자 현황
fig = px.bar(s_df, x='사망자수', y='발생시간', orientation='h')
fig.show()
```

```python
fig = px.bar(s_df, x='발생시간', y='사망자수')
fig.show()
```

```python
# 지도시각화를 위한 라이브러리 임포트
import folium

s_map = folium.Map(location=[37.5666805, 126.9784147])
s_map
```

```python
# 지도시각화를 위한 전처리
s_df = s_df.astype({'사망자수': 'float'})
s_df = s_df.astype({'부상자수': 'float'})
s_df.info()
```

```python
s_df['사고유형'].unique()
```

```python
s_map = folium.Map(location=[37.5666805, 126.9784147])

# s_df 데이터가 있을때까지 반복적으로 CircleMarker를 지도에 add 시키는 작업
for n in s_df.index:
  cnt = s_df['사망자수'][n] + s_df['부상자수'][n]
  # folium.CircleMarker([위도, 경도], 반지름사이즈, 팝업, 색, 채울색).add_to(어느지도에 올릴건지)
  folium.CircleMarker([s_df['위도'][n], s_df['경도'][n]], radius = cnt*10, popup=s_df['사고유형'][n], color='none', fill_color='blue').add_to(s_map)
# color는 line color 의
s_map
```

```python
# 사고 유형별로 횡단중이면 red, 측면충돌이면 green, 길가장자리구역통행중이면 orange, 그 외에는 blue로 그려보자

s_map = folium.Map(location=[37.5666805, 126.9784147])

for n in s_df.index:
  cnt = s_df['사망자수'][n] + s_df['부상자수'][n]

  if s_df['사고유형'][n] == '횡단중' :
    folium.CircleMarker([s_df['위도'][n], s_df['경도'][n]], radius = cnt*10, popup=s_df['사고유형'][n], color='red', fill_color='red').add_to(s_map)

  elif s_df['사고유형'][n] == '측면충돌':
      folium.CircleMarker([s_df['위도'][n], s_df['경도'][n]], radius = cnt*10, popup=s_df['사고유형'][n], color='green', fill_color='green').add_to(s_map)

  elif s_df['사고유형'][n] == '길가장자리구역통행중':
      folium.CircleMarker([s_df['위도'][n], s_df['경도'][n]], radius = cnt*10, popup=s_df['사고유형'][n], color='orange', fill_color='orange').add_to(s_map)
  else:
      folium.CircleMarker([s_df['위도'][n], s_df['경도'][n]], radius = cnt*10, popup=s_df['사고유형'][n], color='blue', fill_color='blue').add_to(s_map)


s_map
```
