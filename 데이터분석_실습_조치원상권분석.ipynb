```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/24-2 데이터분석/데이터셋/세종특별자치시_조치원읍_리별_인구_현황_20221104.csv', encoding='EUC-KR')
df1.head()
```

```python
df1.sort_values(by='세대수', ascending=False, inplace=True, ignore_index=True)
# df1.head(10)/
```

```python
df1.info()
```

```python
df1['구분'].unique()
```

```python
population = df1[['구분', '세대수', '데이터 기준일자']]
population.head()
```

```python
def division(name):
  name = name.split('리')
  if name[0][-1].isdigit():
    return name[0][:-1]+'리'+name[0][-1]
  else:
    return name[0]+'리'+'0'

population['구분'] = population['구분'].apply(division)
population.head()
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/24-2 데이터분석/데이터셋/소상공인시장진흥공단_상가(상권)정보_세종_202409.csv', encoding='UTF-8')
df2.info()
```

```python
jochiwon = df2.iloc[:,[1, 16, 24, 31]].groupby('행정동명').get_group('조치원읍')
jochiwon.drop(columns=['행정동명'], inplace=True)
jochiwon.head(5)
```

```python
population['구분'].unique()
```

```python
def make_lee(input):
  for name in population['구분']:
    if name[:-1] in input:
      return name[:-1]

jochiwon['리'] = jochiwon['지번주소'].apply(make_lee)
jochiwon.head()
```
